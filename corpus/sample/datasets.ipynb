````python
import json
import pandas as pd
from corpus.datasets import DataSets
````

# archived corpus

````python
ds = DataSets()
meta = ds.pull_datasets_meta()

pd.DataFrame(meta).T
````

````python
# columns
pd.DataFrame({x['name']: x['columns'] for x in meta['naver_movie_reviews']['files']}).fillna('-')
````

````python
data = ds.load(
    meta=meta,
    name='naver_movie_reviews', 
    filename='naver.reviews.json.bz2', 
)
print(f'size: {len(data):,}')
````

````python
df = pd.DataFrame(data).fillna('')
df.head()
````

````python
es_meta = ds.pull_elasticsearch_meta()
pd.DataFrame(es_meta).T
````

````python
# columns

pd.DataFrame([es_meta['crawler-naver-economy-2021']['columns']]).T
````
