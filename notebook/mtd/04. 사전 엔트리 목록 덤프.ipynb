```python
import pandas as pd
from utils.elasticsearch_utils import ElasticSearchUtils
```

```python
def save_data(index, data_list):
    df = pd.DataFrame(data_list)

    df.fillna('', inplace=True)

    df.to_json(
        '{}.json.bz2'.format(index),
        force_ascii=False,
        compression='bz2',
        orient='records',
        lines=True,
    )

    return df
```

```python
host_info = {
    'host': 'https://corpus.ncsoft.com:9200',
    'http_auth': 'elastic:nlplab',
}

utils = ElasticSearchUtils(**host_info)
```

```python
index = 'crawler-naver-dictionary'

result = []
utils.export(index=index, result=result)

df = save_data(index=index, data_list=result)
df

# df = pd.DataFrame(result)
# df.fillna('', inplace=True)
# df
```
