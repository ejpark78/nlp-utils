```python
import pandas as pd
```

```python
def insert_data(df):
    """ """
    from uuid import uuid4
    from utils.elasticsearch_utils import ElasticSearchUtils

    host = 'https://corpus.ncsoft.com:9200'
    index = 'corpus-evaluation'

    utils = ElasticSearchUtils(
        host=host, 
        index=index, 
        http_auth='elastic:nlplab'
    )

    for i, row in df.iterrows():
        doc = dict(row)

        doc['_id'] = str(uuid4())
        utils.save_document(document=doc, index=index)

    utils.flush()
    return
```

```python
filename = 'evaluation/200205_L2M_chat_trans_KOEN_1000.xlsx'

alias = {
    '원문': 'korean',
    '번역결과수정': 'english',
    '번역결과': 'tst',
    '번역 방법': 'tst_method',
    '정규화 결과': 'tm_pattern'
}

df = pd.read_excel(filename).fillna('')

df = df.rename(columns=alias)

df.at[:, 'domain'] = 'chatting_log'
df.at[:, 'direction'] = 'korean-english'
df.at[:, 'game_name'] = 'L2M'
df.at[:, 'tag'] = '200205_L2M_chat_trans'

insert_data(df=df)

df
```
