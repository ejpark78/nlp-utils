```python
import re
import pandas as pd
from tqdm.autonotebook import tqdm
```

```python
def read_bitext():
    from glob import glob

    data = {}
    for filename in glob('dictionary/bitext.*'):
        col = filename.replace('dictionary/bitext.', '')
        col = col.replace('.', '_')
        if col not in data:
            data[col] = []

        with open(filename, 'r') as fp:
            for line in fp.readlines():
                line = line.strip()
                line = re.sub('^"(.+)"$', '\g<1>', line)
                line = line.replace('""', '')
                line = line.replace('\\', '')
                line = line.strip()
                line = re.sub('\s+', ' ', line)

                data[col].append(line)

    return pd.DataFrame(data)

df = read_bitext()
df.head()
```

```python
df = df[ df['korean'].str.contains('[ㄱ-ㅣ가-힣]+', regex= True, na=False) ]
df
```

```python
def split_train_test(result_df, df):
    if len(df) == 0:
        return
    
    random_seed = 10

    train_df = df.sample(frac=0.9, random_state=random_seed)
    
    tmp_df = df.drop(train_df.index)
    
    test_df = tmp_df.sample(frac=0.5, random_state=random_seed)
    valid_df = tmp_df.drop(test_df.index)    
    
    result_df.at[test_df.index, 'role'] = 'test'
    result_df.at[valid_df.index, 'role'] = 'dev'
    result_df.at[train_df.index, 'role'] = 'train'
    
    return

def split_data_set(df):
    for theme in set(df['theme'].to_list()):
        for level in set(df['level'].to_list()):
            split_train_test(result_df=df, df=df[(df['theme'] == theme) & (df['level'] == level)])    
    
    return df

df = split_data_set(df)
```

```python
def save_dataset(role):
    with open('dictionary/{role}.en-ko'.format(role=role), 'w') as fp:
        data_df = df[df['role'] == role]
        for i, row in tqdm(data_df.iterrows(), total=len(data_df)):
            line = '{english}\t{korean}\t{theme}\t{level}\n'.format(
                english=row['english_tok'], 
                korean=row['korean_morp_tok'],
                theme=row['theme'],
                level=row['level'],
            )
            fp.write(line)

    return


save_dataset(role='train')
save_dataset(role='dev')
save_dataset(role='test')
```
