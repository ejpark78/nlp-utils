```python
import pandas as pd
```

```python
df = pd.read_json('crawler-ted-2020-07-17.bz2', compression='bz2', lines=True).fillna('').drop(columns=['document_id'])
df
```

```python
lang_list = ['zh-tw', 'vi', 'ko', 'id', 'zh-cn', 'ja', 'en']
```

```python
lang_size = {}
for l in lang_list:
    for l2 in lang_list:
        if l not in lang_size:
            lang_size[l] = {}
            
        lang_size[l][l2] = '{:,}'.format(len(df[ (df[l] != '') & (df[l2] != '') ]))
```

```python
pd.DataFrame(lang_size).fillna(0)
```
