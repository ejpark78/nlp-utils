```python
import pandas as pd
```

```python
df = pd.read_json(
    'crawler-naver-economy-2020.json.bz2',
    compression='bz2',
    orient='records',
    lines=True,
)
```

```python
df.fillna('', inplace=True)
```

```python
df.columns
```

```python
len(df)
```

```python
empty_content = df[df['content'] == '']
len(empty_content)
```

```python
pd.set_option('display.max_rows', None)
empty_content.groupby(by='category').size().to_frame()
```

```python
content = df[df['content'] != '']
len(content)
```

```python
pd.set_option('display.max_rows', None)
content.groupby(by='category').size().to_frame()
```
