```python
def dump_docs(index):
    host_info = {
        'host': 'https://corpus.ncsoft.com:9200',
        'http_auth': 'elastic:nlplab',
    }

    utils = ElasticSearchUtils(**host_info)

    query = {
        '_source': [
            'document_id',
            'date',
            'nlu_wrapper.*.morp_str',
        ]
    }

    doc_list = []
    utils.export(index=index, query=query, result=doc_list)

    for doc in tqdm(doc_list):
        if 'nlu_wrapper' not in doc:
            continue

        nlu_wrapper = doc['nlu_wrapper']

        for k in nlu_wrapper:
            nlu_wrapper[k] = [v['morp_str'] for v in nlu_wrapper[k]]

        if 'title' in nlu_wrapper:
            nlu_wrapper['title'] = '\n'.join(nlu_wrapper['title'])

    with open('data/{}.json'.format(index), 'w') as fp:
        for doc in tqdm(doc_list, desc=index):
            line = json.dumps(doc, ensure_ascii=False) + '\n'
            fp.write(line)

    return
```
