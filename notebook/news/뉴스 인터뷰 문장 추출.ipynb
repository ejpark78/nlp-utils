# 댓글 정보 추출 

```python
def export_reply(host, auth, index_list):
    """뉴스 본문이나 댓글을 덤프 받는다."""
    from utils.elasticsearch_utils import ElasticSearchUtils

    elastic = ElasticSearchUtils(host=host, http_auth=auth)

    query = {
        '_source': [
            'aid', 'oid', 'sectionName', 'officeName', 'title',
            'reply_list.contents', 'reply_list.userName', 'reply_list.regTime', 'reply_list.maskedUserId'
        ]
    }

    result = []
    for index in index_list:
        elastic.export(
            index=index,
            query=query,
            result=result,
        )

    return result
```

```python
host = 'https://corpus.ncsoft.com:9200'
auth = 'crawler:crawler2019'

index_list = ['crawler-naver-sports-reply-2018', 'crawler-naver-sports-reply-2019']

reply_list = export_reply(host, auth, index_list)
```

```python
coach = clean_df[clean_df['who'].str.contains('감독')]
len(coach)
```

```python
reply_idx = {}
for r in reply_list:
    if 'reply_list' not in r:
        continue

    doc_id = '{oid}-{aid}'.format(**r)    
    reply_idx[doc_id] = len(r['reply_list'])
```

```python
list(reply_idx.keys())[:10]
```

```python
count = 0
for i, row in coach.iterrows():
    doc_id = row['doc_id']
    if doc_id not in reply_idx:
        continue
        
    count += reply_idx[doc_id]
    
count
```
