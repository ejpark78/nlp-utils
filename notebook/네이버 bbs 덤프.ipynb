```python
import numpy as np
import pandas as pd

from tqdm import tqdm
from utils.elasticsearch_utils import ElasticSearchUtils
```

```python
def dump_data(data_id, field):
    """ """
    index = 'crawler-bbs-naver'
    
    host_info = {
        'host': 'https://corpus.ncsoft.com:9200',
        'http_auth': 'elastic:nlplab',
    }

    utils = ElasticSearchUtils(**host_info)

    query = {
        'query': {
            'bool': {
                'must': {
                    'match': {
                        field: data_id
                    }
                }
            }
        }
    }

    doc_list = []
    utils.export(index=index, query=query, result=doc_list)

    return doc_list
```

```python
doc_list = dump_data(data_id=29846718, field='clubid')

'{:,}'.format(len(doc_list))
```

```python
df = pd.DataFrame(doc_list)

df.fillna('', inplace=True)
    
print('{:,}'.format(len(df)))
df.head()
```

```python
field = 'bbs_name'

target_df = df[ df[field].str.contains('제작기') ]
target_df
```

```python
from os.path import isfile
from weasyprint import HTML

for i, row in tqdm(target_df.iterrows(), total=len(target_df)):
    filename = 'data/blog/쿠치레/제작기/{}-{}.html'.format(row['document_id'], row['title'].replace('쿠치레 ', ''))
    if isfile(filename) is True:
        continue
    
    with open(filename, 'w') as fp:
        fp.write(row['html_content'])
        
    pdf_filename = filename.replace('.html', '.pdf')
    HTML(string=row['html_content']).write_pdf(pdf_filename)
```
