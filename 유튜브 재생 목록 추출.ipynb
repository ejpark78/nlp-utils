```python
from tqdm import tqdm

for i, row in tqdm(df.iterrows(), total=len(df)):
    title = row['title']
    if title not in title_index:
        continue
        
    doc = {
        '_id': row['document_id'],
        'streamer': '똘끼100%'
    }    
    elastic.save_document(document=doc, delete=False)
    
elastic.flush()
```
